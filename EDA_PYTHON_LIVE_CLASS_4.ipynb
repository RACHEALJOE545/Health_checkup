```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
Medic_data["sex"] = encoder.fit_transform(Medic_data["sex"])
Medic_data.head()
```

```python
Medic_data["region"] = encoder.fit_transform(Medic_data["region"])
Medic_data["smoker"] = encoder.fit_transform(Medic_data["smoker"])
Medic_data.head()
```

```python
correlation = Medic_data.corr()
correlation
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(correlation, annot = True, fmt= ".2f", cmap = "Blues")
plt.title("Medical Client Correlation")
plt.show()
```
